Na aula de hoje, vamos trabalhar com **geoprocessamento**, que nada mais é do que a exibição de dados em mapas geográficos. Na célula de código abaixo, vamos importar a biblioteca de geração de mapas geográficos.Para isso, vamos fazer uso da biblioteca **folium**. Para usarmos, precisaremos instalá-la atravéz do comando `pip install folium`.


```python
import folium
```

Na próxima célula de código, vamos gerar um mapa e fazê-lo abrir na cidade de **Brasília-DF**

```python
mapa = folium.Map(location = [-15.7834301996, -47.899],zoom_start=13 )
display(mapa)
```

Na próxima célula de código vamos criar marcações de lugares.

```python
mapa = folium.Map(location = [-15.7834301996, -47.899],zoom_start=15 )
folium.Marker(location=[-15.7948, -47.91319], popup="SENAI SIG. \nOnde está acontecendo o curso de Python Developer!").add_to(mapa)

folium.Marker(location=[-15.7834301996, -47.899], popup=" Estádio Mane Garrincha.", icon=folium.Icon(color='red')).add_to(mapa)
display(mapa)
```

Salvando o mapa em uma página HTML:

```python
mapa.save("index.html")
```
